## Tools
In this lab we will make use of:

- Matplotlib, a popular library for plotting data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the function f_{w,b}(x)
def f_wb(x, w, b):
    return w * x + b

# Define the cost function J(w, b)
def J(w, b, x, y):
    m = len(x)  # Number of training examples. In our lecture example, m=3
    total_cost = 0
    for i in range(m):
        prediction = f_wb(x[i], w, b)
        error = prediction - y[i]
        total_cost += error ** 2
    return (1 / (2 * m)) * total_cost
```

```python
# Define the lecture dataset
X = np.array([1, 2, 3])
y = np.array([1, 2, 3])

# Create the scatter plot
plt.scatter(X, y, color='red', marker='x', s=80)

# Generate x values for the function plot
x_values = np.linspace(0, 4, 100)  # Generate 100 points between 0 and 4

# Set b=0 (For simplicity)
b = 0

w_numbers = [1, 0.5, 0, -0.5]
for w in w_numbers:
    y_values = f_wb(x_values, w, b)
    plt.plot(x_values, y_values, label=f'f(x) = {w}x')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Plot of f(x)=wx. Consider b=0')
plt.legend()

# Show the plot
plt.show()
```

```python
w_scatter   = []
J_w_scatter = []

# Lecture order for w
#w_values_to_plot = [0, 0.5, 1, -0.5, 1.5, 2, 2.5, 1.25, 0.75, 1.75, 0.25, 2.25, -0.25]

# Ordered list of values to plot a continuous line
w_values_to_plot = [-0.5, -0.25, 0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2, 2.25, 2.5]
for w in w_values_to_plot:
    cost = J(w, b, X, y)
    cost = round(cost, 4)
    w_scatter.append(w)
    J_w_scatter.append(cost)
    print(f"For b = {0} and w = {w:<5}, Cost J(w): {cost}")
```

```python
# Create the scatter plot
plt.scatter(w_scatter, J_w_scatter, color='red', marker='x', s=80)

# Plot a continuous line
plt.plot(w_scatter, J_w_scatter, color='black', label='Continuous line')

# Add labels and legend
plt.xlabel('w')
plt.ylabel('J(w)')
plt.title('Plot of J(w)')
plt.legend()

# Show the plot
plt.show()
```

## Final code in a cell

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the lecture dataset
X = np.array([1, 2, 3])
y = np.array([1, 2, 3])

# Define the function f_{w,b}(x)
def f_wb(x, w, b):
    return w * x + b

# Define the cost function J(w, b)
def J(w, b, x, y):
    m = len(x)  # Number of training examples. In our lecture example, m=3
    total_cost = 0
    for i in range(m):
        prediction = f_wb(x[i], w, b)
        error = prediction - y[i]
        total_cost += error ** 2
    return (1 / (2 * m)) * total_cost

# Create the scatter plot
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Plot f(x)=wx
axs[0].scatter(X, y, color='red', marker='x', s=80)
x_values = np.linspace(0, 4, 100)  # Generate 100 points between 0 and 4
b = 0
w_numbers = [1, 0.5, 0, -0.5]
for w in w_numbers:
    y_values = f_wb(x_values, w, b)
    axs[0].plot(x_values, y_values, label=f'f(x) = {w}x')
    
axs[0].set_xlabel('x')
axs[0].set_ylabel('y')
axs[0].set_title('Plot of f(x)=wx. Consider b=0')
axs[0].legend()

# Plot J(w)
w_scatter = []
J_w_scatter = []
w_values_to_plot = [-0.5, -0.25, 0, 0.25, 0.5, 0.75, 1, 1.25, 1.5, 1.75, 2, 2.25, 2.5]
for w in w_values_to_plot:
    cost = J(w, b, X, y)
    cost = round(cost, 4)
    w_scatter.append(w)
    J_w_scatter.append(cost)

axs[1].scatter(w_scatter, J_w_scatter, color='red', marker='x', s=80)
axs[1].plot(w_scatter, J_w_scatter, color='black', label='Continuous line')
axs[1].set_xlabel('w')
axs[1].set_ylabel('J(w)')
axs[1].set_title('Plot of J(w)')
axs[1].legend()

plt.tight_layout()
plt.show()
```
