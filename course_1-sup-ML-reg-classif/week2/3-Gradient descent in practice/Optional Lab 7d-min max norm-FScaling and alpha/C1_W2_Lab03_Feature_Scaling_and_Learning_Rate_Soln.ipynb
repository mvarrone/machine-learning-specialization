## Gradient Descent With Multiple Variables
Here are the equations you developed in the last lab on gradient descent for multiple variables.:

$$\begin{align*} \text{repeat}&\text{ until convergence:} \; \lbrace \newline\;
& w_j := w_j -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial w_j} \tag{1}  \; & \text{for j = 0..n-1}\newline
&b\ \ := b -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial b}  \newline \rbrace
\end{align*}$$

where, n is the number of features, parameters $w_j$,  $b$, are updated simultaneously and where  

$$
\begin{align}
\frac{\partial J(\mathbf{w},b)}{\partial w_j}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)})x_{j}^{(i)} \tag{2}  \\
\frac{\partial J(\mathbf{w},b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}) \tag{3}
\end{align}
$$
* m is the number of training examples in the data set

    
*  $f_{\mathbf{w},b}(\mathbf{x}^{(i)})$ is the model's prediction, while $y^{(i)}$ is the target value

The figure above shows why $w$'s are updated unevenly. 
- $\alpha$ is shared by all parameter updates ($w$'s and $b$).
- the common error term is multiplied by the features for the $w$'s. (not $b$).
- the features vary significantly in magnitude making some features update much faster than others. In this case, $w_0$ is multiplied by 'size(sqft)', which is generally > 1000,  while $w_1$ is multiplied by 'number of bedrooms', which is generally 2-4. 
    
The solution is Feature Scaling.

The lectures discussed three different techniques on Feature Scaling: 
- Max normalization: Essentially, dividing each positive feature by its maximum value (explored in Lab 7b)
- Mean normalization: $x_i := \dfrac{x_i - \mu_i}{max(x) - min(x)}$ Features will have a range from -1 to 1 (explored in Lab 7c)
- Z-score normalization which we explored in a previous lab (explored in Lab 7a)
- Min Max normalization: $x_i := \dfrac{x_i - min(x)}{max(x) - min(x)}$ Features will have a range from 0 to 1 (Objective of this Lab 7d)


### Min Max normalization 
After Min Max normalization, all features will have a range from 0 to 1

To implement Min Max normalization, adjust your input values as shown in this formula:
$$x_j := \dfrac{x_j - min(x)}{max(x) - min(x)}$$
where $j$ selects a feature or a column in the $\mathbf{X}$ matrix

>**Implementation Note:** When normalizing the features, it is important
to store the values used for normalization - the min and max values are used for the computations. After learning the parameters
from the model, we often want to predict the prices of houses we have not
seen before. Given a new x value (living room area and number of bed-
rooms), we must normalize x using the calculated min and max values that we had previously computed from the training set.

**Implementation**

```python
def min_max_normalize_features(X):
    """
    computes  X, min max normalized by column
    
    Args:
      X (ndarray (m,n))     : input data, m examples, n features
      
    Returns:
      X_norm (ndarray (m,n)): input normalized by column
      X_maximum
      X_minimum
    """

    # Find the max of each column/feature
    X_maximum = np.max(X, axis=0)     

    # Find the min of each column/feature
    X_minimum = np.min(X, axis=0)     

    # element-wise, subtract X_minimum for that column from each example and then divide by difference between the max and the min for that column
    X_norm = (X - X_minimum) / (X_maximum - X_minimum)      

    return (X_norm, X_maximum, X_minimum)
```

Let's look at the steps involved in min max normalization. The plot below shows the transformation step by step.

```python
maximum = np.max(X_train, axis=0) 
minimum = np.min(X_train, axis=0) 

X_norm = (X_train - minimum) / (maximum - minimum)      

fig,ax = plt.subplots(1, 2, figsize=(12, 3))

# Before normalization graph
ax[0].scatter(X_train[:,0], X_train[:,3])
ax[0].set_xlabel(X_features[0]); ax[0].set_ylabel(X_features[3]);
ax[0].set_title("Unnormalized")
ax[0].axis('equal')

# After normalization graph
ax[1].scatter(X_norm[:,0], X_norm[:,3])
ax[1].set_xlabel(X_features[0]); ax[0].set_ylabel(X_features[3]);
ax[1].set_title(r"Min Max normalized")
ax[1].axis('equal')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
fig.suptitle("Distribution of features before and after normalization")
plt.show()
```

The plot above shows the relationship between two of the training set parameters, "age" and "size(sqft)". *These are plotted with equal scale*. 
- Left: Unnormalized: The range of values or the variance of the 'size(sqft)' feature is much larger than that of age
- Right: The second step implements a difference with minimun and then divides by the difference between the max and the minimum. This leaves both features with a similar scale and a peak to peak value of 1. Ranges are from 0 to 1

Let's normalize the data and compare it to the original data.

```python
# normalize the original features
X_norm, X_maximum, X_minimum = min_max_normalize_features(X_train)

print(f"X_maximum = {X_maximum}\nX_minimum = {X_minimum}")
raw_ptp = np.ptp(X_train,axis=0)
norm_ptp = np.ptp(X_norm,axis=0)
print(f"Peak to Peak range by column in Raw        X:{raw_ptp}")   
print(f"Peak to Peak range by column in Normalized X:{norm_ptp}")
```

The peak to peak range of each column is reduced from a factor of thousands to a factor of by normalization.

```python
# Distribution of features before normalization
fig, ax = plt.subplots(1, 4, figsize = (12, 3))
for i in range(len(ax)):
    norm_plot(ax[i],X_train[:,i],)
    ax[i].set_xlabel(X_features[i])

ax[0].set_ylabel("count")
fig.suptitle("Distribution of features before normalization")
plt.show()

# Distribution of features after normalization
fig, ax = plt.subplots(1,4,figsize = (12,3))
for i in range(len(ax)):
    norm_plot(ax[i],X_norm[:,i],)
    ax[i].set_xlabel(X_features[i])

ax[0].set_ylabel("count")
fig.suptitle("Distribution of features after normalization")
plt.show()
```

## Printing feature ranges before and after Normalization

```python
X_features = ['size (sqft)', 'bedrooms', 'floors', 'age']

# Function to convert integer to subscript
def subscript(i):
    subscripts = ""
    return ''.join(subscripts[int(digit)] for digit in str(i))

def print_feature_ranges(X_features, X):
    for i in range(len(X_features)):
        range_minimum, range_maximum = X[:, i].min(), X[:, i].max()
        range_average = round((range_minimum + range_maximum) / 2, 2)
        print(f"{X_features[i]:<12}: {range_minimum:>6.2f} <= x{subscript(i)} <= {range_maximum:>7.2f} || x{subscript(i)} Range average value: {range_average} ")
```

```python
print("Before Min Max Normalization\n")
print_feature_ranges(X_features=X_features, X=X_train)
print(f"\nPeak to Peak range by column in Raw: {raw_ptp}")   
```

```python
print("After Min Max Normalization\n")
print_feature_ranges(X_features=X_features, X=X_norm)
print(f"\nPeak to Peak range by column in Normalized X: {norm_ptp}")
```

Notice, above, the range of the normalized data (x-axis) is centered around 0.5 and roughly +/- 0.5. Most importantly, the range is similar for each feature.

Let's re-run our gradient descent algorithm with normalized data.
Note the **vastly larger value of alpha**. This will speed up gradient descent.

```python
w_norm, b_norm, hist = run_gradient_descent(X=X_norm, y=y_train, iterations=1000, alpha=1.0e-1)
```

The scaled features get very accurate results **much, much faster!**. Notice the gradient of each parameter is tiny by the end of this fairly short run. A learning rate of 0.1 is a good start for regression with normalized features.
Let's plot our predictions versus the target values. Note, the prediction is made using the normalized feature while the plot is shown using the original feature values.

```python
# Predict target using normalized features
m = X_norm.shape[0]
yp = np.zeros(m)
for i in range(m):
    yp[i] = np.dot(X_norm[i], w_norm) + b_norm

# Plot predictions and targets versus original features    
fig,ax=plt.subplots(1,4,figsize=(12, 3),sharey=True)
for i in range(len(ax)):
    ax[i].scatter(X_train[:,i],y_train, label = 'Target')
    ax[i].set_xlabel(X_features[i])
    ax[i].scatter(X_train[:,i],yp,color=dlc["dlorange"], label = 'Predict')
    
ax[0].set_ylabel("Price")
ax[0].legend()
fig.suptitle("Target vs Prediction using Min Max normalized model")
plt.show()
```

The results look good. A few points to note:
- with multiple features, we can no longer have a single plot showing results versus features.
- when generating the plot, the normalized features were used. Any predictions using the parameters learned from a normalized training set must also be normalized.

**Prediction**
The point of generating our model is to use it to predict housing prices that are not in the data set. Let's predict the price of a house with 1200 sqft, 3 bedrooms, 1 floor, 40 years old. Recall, that you must normalize the data with the min and max derived when the training data was normalized. 

```python
# First, normalize our example.
x_house = np.array([1200, 3, 1, 40])
x_house_norm = (x_house - X_minimum) / (X_maximum - X_minimum)
print("x_house_norm = ", x_house_norm)

x_house_predict = np.dot(x_house_norm, w_norm) + b_norm
print(f"The predicted price of a house with {x_house[0]} sqft, {x_house[1]} bedrooms, {x_house[2]} floor, {x_house[3]} years old = $ {x_house_predict * 1000:0.0f}")
```
